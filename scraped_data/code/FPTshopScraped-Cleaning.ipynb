```python
import pandas as pd
```

```python
data = pd.read_csv('./result_detail_final.csv')
data.head()
```

```python
data[['Brand', 'Series']] = data.name.str.split(n = 1, expand=True)
data = data.drop('name', 1)
data.head()
```

```python
data['RAM'] = data['RAM'].str.split(' ', 1).str[0]
data.head()
```

```python
data[['storage_type', 'storage_size']] = data['hard_drive'].str.split(' ', n = 1, expand = True)
data = data.drop('hard_drive', 1)
data.head()
```

```python
data['screen_size'] = data['screen_size'].str.strip('"')
data['screen_size']
```

```python
data['display_resolution'] = data['display_resolution'].str.split(' ', n = 3).str[0:3].str.join(' ')
data['display_resolution'] 
```

```python
data['display_rating'] = data['display_rating'].str.split(' ', 1).str[0]
data['display_rating'] 
```

```python
cols = data.columns.tolist()
print(cols)
```

```python
cols  = ['PN', 'rating', 'rating_count','Brand', 'Series','price', 'CPU', 
         'RAM', 'RAM_upgradable', 'GPU', 'battery_wh', 'storage_type', 'storage_size',
         'storage_upgradable', 'screen_size', 'display_resolution', 'display_rating', 'screen_technology', 
         'weight', 'length', 'width', 'height', 'OS', 'surface_material', 'release_year', 
         'ports', 'keyboard_light']
data = data.reindex(columns = cols)
data.iloc[0,:]
```

```python
data.head()
```

```python
data.to_csv('result_cleaned.csv')
```
